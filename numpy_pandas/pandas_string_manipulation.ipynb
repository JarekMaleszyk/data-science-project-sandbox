```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {
        "email":[
            "john.paul@examplepetstore.com",
            "jimmi.a.hendrix@example.com",
            "jamue.gonzales@gmail.com",
            "me@sara.edu.net",
            "william.henry.moody@my-own-personal-domain.com",
            "william-test@gmail.com",
            "william.henry.harrison@example-pet-store.com",
            "mike@gmail.com"
        ]
    }
)
df
```

```python
df['is_gmail'] = df['email'].str.contains('gmail')
df
```

```python
df['domain'] = df['email'].str.split('@').str[1]
df
```

```python
df['domain_descriptor'] = df['domain'].str.replace('.', '_').str.replace('-', '_').str.lower()
df
```

```python
df['name'] = df['email'].str.split('@').str[0].str.title()
df
```

```python
df['example'] = df['domain_descriptor'].str.startswith('example')
df
```

```python
df['email_length'] = df['email'].str.len()
df
```

```python
df['domain_descriptor'] = df['domain_descriptor'].str.strip()
df
```
